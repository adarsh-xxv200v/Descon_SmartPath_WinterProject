```python
# Task1
cityNames = ["Mumbai", "New York", "Rio", "Philippines", "Dubai", "Delhi"]
numberRoads = [456,878,323,464,321,566]
averageSpeed = [34,54,23,67,87,23]

# Task2
for i in range(6):
    print(f"{cityNames[i]} has {numberRoads[i]} number of roads and average speed there is {averageSpeed[i]}")

print()

count = 0
lst = []
for i in range(6):
    if averageSpeed[i]<=30:
        lst.append(cityNames[i])
        count += 1
print("Number of cities with traffic: ",count)
print("The cities: ", lst)
```

```python
#Task3
import numpy as np
arr = np.array(averageSpeed)
print("Average speed: ", np.mean(arr))
print("Maximum speed: ",arr[arr.argmax()])
print("Minimum speed: ",arr[arr.argmin()])

for i in range(6):
    print("Average time in minutes required for travelling 2 km in ", cityNames[i], ":", 2*60/averageSpeed[i])
```

```python
#Task4
import pandas as pd
dataSet = {
    "Road_ID": [10,12,24,6,13],
    "Speed": [67,32,89,56,21],
    "Length": [456,546,678,232,760]
}

df = pd.DataFrame(dataSet)
df["Travel_Time"] = [(dataSet["Length"][i] / dataSet["Speed"][i]) for i in range(5)]
df["Traffic_Level"] = ["Low", "High", "Medium", "Low", "Severe"]
df.sort_values("Travel_Time", ascending=False, inplace=True)
print(df)
```

```python
#Task5
import matplotlib.pyplot as plt
plt.plot(df["Road_ID"],df["Travel_Time"])
plt.xlabel("Road_ID")
plt.ylabel("Travel_Time")
plt.grid()
plt.title("Line chart")
plt.show()
```

```python
plt.bar(df["Road_ID"],df["Travel_Time"])
plt.xlabel("Road_ID")
plt.ylabel("Travel_Time")
plt.grid()
plt.title("Bar chart")
plt.show()
```

```python
#Bonus task
plt.scatter(df["Speed"],df["Travel_Time"])
plt.xlabel("Speed")
plt.ylabel("Travel_Time")
plt.grid()
plt.title("Scatter chart")
plt.show()
```
